```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import statsmodels.api as sm
import statsmodels.tsa.api as smt
import statsmodels.tsa.arima.model as smtar

from scipy.stats import boxcox

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import TimeSeriesSplit

from statsmodels.tsa.stattools import adfuller

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
robberies_in_boston = pd.read_csv("Series/monthly-boston-armed-robberies-j.csv")
robberies = robberies_in_boston["Count"]
```

```python
def test_stationarity(timeseries):
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for [key, value] in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)
```

```python
def tsplot(y, lags=None, figsize=(14, 8), style='bmh'):
    test_stationarity(y)
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):
        plt.figure(figsize=figsize)
        layout = (5, 1)
        ts_ax = plt.subplot2grid(layout, (0, 0), rowspan=2)
        acf_ax = plt.subplot2grid(layout, (2, 0))
        pacf_ax = plt.subplot2grid(layout, (3, 0))
        qq_ax = plt.subplot2grid(layout, (4, 0))

        y.plot(ax=ts_ax, color='blue', label='Or')
        ts_ax.set_title('Original')

        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.05)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.05)
        sm.qqplot(y, line='s', ax=qq_ax)
        
        plt.tight_layout()
    return
```

```python
tsplot(robberies)
```

```python
mse = dict.fromkeys(['ARIMA', 'LR', 'LR_with_ARIMA', 'RF', 'RF_with_ARIMA'])
```

```python
%%time
# Подбираю параметры для ARIMA через максимизацию aic

best_aic = np.inf 
best_order = None
best_mdl = None
best_method = None

for i in range(5):
    for d in range(5):
        for j in range(5):
            for m in ['statespace', 'innovations_mle', 'hannan_rissanen', 'burg', 'innovations', 'yule_walker']:
                try:
                    tmp_mdl = smtar.ARIMA(robberies, order=(i, d, j)).fit(method=m)
                    tmp_aic = tmp_mdl.aic
                    if tmp_aic < best_aic:
                        best_aic = tmp_aic
                        best_order = (i, d, j)
                        best_mdl = tmp_mdl
                        best_method = m
                except: continue

mse['ARIMA'] = best_mdl.mse
print(f'aic: {best_aic:6.5f} | order: {best_order} | method: {best_method}')
print(f'mse: {best_mdl.mse:6.5f}\n')

tsplot(best_mdl.resid, lags=30)
```

Остатки распределены нормально, т. е. модель отработала хорошо. При этом mse довольно большое и на графике видно, что предсказание не идеально, хоть и учитывает тренд.

```python
print(f'MSE: {best_mdl.mse:5f}')

with plt.style.context('bmh'):
    plt.figure(figsize=(14,8))
    ax = plt.axes()
    plt.plot(best_mdl.predict(1, len(robberies)+30))
    plt.plot(robberies, color='red', label='Series')
    plt.legend()
    plt.title('ARIMA')
    plt.show()
```

Построю модели линейной регрессии и Random Forest в двух вариантах: с включением в качестве признака результатов работы ARIMA и без этого признака.

```python
def prepareData(data, lag_start=5, lag_end=20, test_size=0.15, ARIMA=False):
    
    data = pd.DataFrame(data.copy())
    
    # считаем индекс в датафрейме, после которого начинается тестовый отрезок
    test_index = int(len(data)*(1-test_size))
    
    # добавляем лаги исходного ряда в качестве признаков
    for i in range(lag_start, lag_end):
        data["lag_{}".format(i)] = data.Count.shift(i)
    
    if ARIMA:
        data['ARIMA'] = best_mdl.predict()
    
    data = data.dropna()
    data = data.reset_index(drop=True)
    data = data.drop(["Month"], axis=1)
     
    # разбиваем весь датасет на тренировочную и тестовую выборку
    X_train = data.loc[:test_index].drop(["Count"], axis=1)
    y_train = data.loc[:test_index]["Count"]
    X_test = data.loc[test_index:].drop(["Count"], axis=1)
    y_test = data.loc[test_index:]["Count"]
    
    return X_train, X_test, y_train, y_test
```

```python
#Функия для получения прогнозных значений
def make_prediction(data, steps, model, lag_start, lag_end, ARIMA=False):
    prediction = list(data)
    
    if ARIMA:
        arlist = best_mdl.predict(0, len(data) + steps)
    
    for i in range(steps):
        if not ARIMA:
            X_future = np.array(prediction[-(lag_start):-(lag_end):-1]).reshape(1, -1)
        else:
            X_future = np.array(prediction[-(lag_start):-(lag_end):-1] + [arlist[len(data) + i + 1]]).reshape(1, -1)
        prediction.append(float(model.predict(X_future)))
    
    return prediction
```

```python
lag_start = 1
lag_end = 30
X_train, X_test, y_train, y_test = prepareData(robberies_in_boston, lag_start=lag_start, lag_end=lag_end, test_size=0.3)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
prediction = lr.predict(X_test)

plt.figure(figsize=(14, 8))
plt.plot(make_prediction(robberies, 30, lr, lag_start=lag_start, lag_end=lag_end), "r", label="prediction")
plt.plot(robberies, label="actual")
plt.legend(loc="best")
plt.title("Linear regression")
plt.grid(True)
print(f'MSE: {mean_squared_error(y_test, prediction):5f}')

mse['LR'] = mean_squared_error(y_test, prediction)
```

Несмотря на большую ошибку, предсказание выглядит правдоподобным.

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
prediction = rf.predict(X_test)

plt.figure(figsize=(14, 8))
plt.plot(make_prediction(robberies, 50, rf, lag_start=lag_start, lag_end=lag_end), "r", label="prediction")
plt.plot(robberies, label="actual")
plt.legend(loc="best")
plt.title("Random forest")
plt.grid(True)

print(f'MSE: {mean_squared_error(y_test, prediction):5f}')

mse['RF'] = mean_squared_error(y_test, prediction)
```

MSE гораздо меньше, чем у линейной регрессии, но при этом график очень быстро затухает.

```python
# Теперь то же самое плюс данные ARIMA
log_start = 1
log_end = 30
X_train, X_test, y_train, y_test = prepareData(robberies_in_boston, lag_start=log_start, lag_end=log_end, test_size=0.3, ARIMA=True)
```

```python
lra = LinearRegression()
lra.fit(X_train, y_train)
prediction = lra.predict(X_test)

plt.figure(figsize=(14, 8))
plt.plot(make_prediction(robberies, 50, lra, lag_start=lag_start, lag_end=lag_end, ARIMA=True), "r", label="prediction")
plt.plot(robberies, label="actual")
plt.legend(loc="best")
plt.title("Linear regression_with_ARIMA")
plt.grid(True)
print(f'MSE: {mean_squared_error(y_test, prediction):5f}')

mse['LR_with_ARIMA'] = mean_squared_error(y_test, prediction)
```

Добавление данных ARIMA позволило немного уменьшить ошибку, но все равно она гигантская.

```python
rfa = RandomForestRegressor()
rfa.fit(X_train, y_train)
prediction = rfa.predict(X_test)

plt.figure(figsize=(14, 8))
plt.plot(make_prediction(robberies, 50, rfa, lag_start=lag_start, lag_end=lag_end, ARIMA=True), "r", label="prediction")
plt.plot(robberies, label="actual")
plt.legend(loc="best")
plt.title("Random forest_with_ARIMA")
plt.grid(True)

print(f'MSE: {mean_squared_error(y_test, prediction):5f}')

mse['RF_with_ARIMA'] = mean_squared_error(y_test, prediction)
```

По сравнению с предыдущим Random Forest особенно ничего не изменилось.

```python
print('MSE для всех использованных моделей\n')
for key in sorted(mse, key=lambda x: mse[x]):
    print(f'{key:>14}: {mse[key]:10.5f}')
```

Линейная регрессия предсказывает данные с большой ошибкой, даже если ей передать результаты ARIMA. При этом предсказания. судя по графику, получаются правдоподобными.

RandomForest справился с меньшей ошибкой, но его прогноз постепенно сводится к константе.Результаты Random Forest при добавлении столбца ARIMA улучшаются незначительно (при некоторых запусках RF показывал MSE меньше, чем RF_with_ARIMA)
