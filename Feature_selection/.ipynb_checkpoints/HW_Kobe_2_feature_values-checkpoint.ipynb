```python
from sklearn.decomposition import PCA
from sklearn.linear_model import Lasso
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import RidgeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler

import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

Итак, я нагенерила огромное количество признаков. Теперь надо посмотреть на их корреляции и на то, как они работают в различных моделях

```python
data = pd.read_excel('data_plus.xlsx')
```

```python
del data['Unnamed: 0']
data.head()
```

```python
data.info()
```

Убираю все столбцы с нечисловым типом данных, все id, все IV

```python
columns = set(data) - {'action_type', 'combined_shot_type', 'game_event_id', 'game_id', 'lat', 'loc_x', 'loc_y', 'lon',
                      'minutes_remaining', 'season', 'seconds_remaining', 'shot_made_flag', 'shot_type',
                      'shot_zone_area', 'shot_zone_basic', 'shot_zone_range', 'team_id', 'team_name', 'game_date', 'matchup',
                      'opponent', 'shot_id', 'IV_combined_shot_type', 'IV_action_type', 'IV_opponent', 'shot_area_basic',
                       'IV_shot_zone_area', 'IV_shot_type', 'IV_shot_zone_basic', 
                       'IV_shot_zone_range', 'IV_shot_area_basic', 'date_as_date'}
columns = list(columns)
len(columns)
```

```python
#Создаю выборку с известными значениями таргетных переменных
data_train = data[~(data.shot_made_flag.isna())]
```

Есть ощущение, что данные сильно скоррелированы между собой. Поэтому нужно посмотреть матрицу корреляций и убрать самые вопиющие случаи. Признаков так много, что рисовать heatmap не имеет смысла.

```python
data_corr = data_train[list(columns) + ['shot_made_flag']].corr()
data_corr
```

```python
corrs = []

for row in columns:
    for col in columns[columns.index(row):]:
        if abs(data_corr.loc[row, col]) > 0.95 and row != col:
            corrs.append((row, col, data_corr.loc[row, col]))
            
sorted(corrs, key=lambda x: abs(x[2]),reverse=True)
```

```python
items_to_remove = ['SAB_Center(C)__Restricted Area', 'SZR_Back Court Shot', 'SAB_Back Court(BC)__Backcourt',
                   'WOE_shot_type', 'shot_distance', 'seasons_code', 'year', 'WOE_shot_area_basic',
                  'period', 'all_seconds_remaining', 'SZR_24+ ft.']

columns = [col for col in columns if col not in items_to_remove]
len(columns)
```

```python
corrs = []

for row in columns:
    for col in columns[columns.index(row):]:
        if abs(data_corr.loc[row, col]) > 0.90 and row != col:
            corrs.append((row, col, data_corr.loc[row, col]))
            
sorted(corrs, key=lambda x: abs(x[2]),reverse=True)
```

```python
items_to_remove = ['WOE_shot_zone_range', 'SZB_Backcourt', 'SZB_Above the Break 3', 'SZB_Restricted Area', 'WOE_shot_zone_basic']

columns = [col for col in columns if col not in items_to_remove]
len(columns)
```

```python
corrs = []

for row in columns:
    for col in columns[columns.index(row):]:
        if abs(data_corr.loc[row, col]) > 0.85 and row != col:
            corrs.append((row, col, data_corr.loc[row, col]))
            
sorted(corrs, key=lambda x: abs(x[2]),reverse=True)
```

```python
columns.remove('CST_Dunk')
len(columns)
```

```python
#Корреляция признаков с таргетной переменной
features = list()

for feature in columns:
    features.append ((feature, data_corr[feature].loc['shot_made_flag']))

for i in sorted(features, key=lambda x: abs(x[1]), reverse=True):
    print(f'{i[0]}: {i[1]}')
```

Оказывается, что с целевой переменной лучше всего коррелируют пространственные признаки. Время, загруженность сезона, оппонент и т.д. оказались в самом конце таблицы с минимальными значениями.

```python
#Создаю таблицу, где буду хранить характеристики переменных
columns_value = pd.DataFrame(data={'correlation':data_corr[columns].loc['shot_made_flag']})
```

Следующий этап: строю линейные модели и смотрю, как зануляются признаки

```python
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(data_train[columns]), columns = columns)
y = data_train['shot_made_flag']
```

```python
#Lasso
alphas=list(2**np.linspace(-10,10,100))
models, preds = [], []

for a in alphas:
    model = Lasso(alpha=a).fit(X, y)
    models.append(model)
    preds.append(model.coef_)
    
plt.figure(figsize=(12, 12))
for i in np.vstack(preds).T:
    plt.plot(range(len(preds)), i)

plt.ylim(-0.05,0.05) 
plt.legend(columns, loc=5)      
plt.grid()
```

Позже всего зануляются признаки с самой высокой корреляцией: WOE_action_type и action_type_Jump Shot. Но в остальном последовательность получается совершенно не такой, как в таблице корреляций: поздно зануляются временные признаки (например, time_behind - время, которое прошло с начала матча). При этом признаки с относительно высокой корреляцией, например eucl_distance, получают нулевой коэффициент уже на второй итерации. Очень странно, что сразу же занулился CST_Jump Shot - один из самых важных признаков по IV.

```python
zeros_count = []

for arr in np.vstack(preds).T:
    zeros_count.append(list(arr).count(0))
    
columns_value['lasso'] = zeros_count

columns_value.sort_values(by='lasso')
```

```python
# ridge
alphas=list(2**np.linspace(15, 35, 100))
models, preds = [], []

for a in alphas:
    model = RidgeClassifier(alpha=a).fit(X, y)
    models.append(model)
    preds.append(model.coef_)
    
plt.figure(figsize=(12, 12))
for i in np.vstack(preds).T:
    plt.plot(range(len(preds)), i)

plt.ylim(-0.04,0.04) 
plt.legend(columns, loc=5)      
plt.grid()
```

Этот график мне сложно проанализировать - ни один коэффициент не приходит к нулю и все приходят к минимуму как-то одновременно. Но WOE_action_type и action_type_jump_shot опять на лидирующих позициях

```python
# Логистическая регрессия
alphas = [2 ** - x for x in np.linspace(1, 25, 100)]

models, preds = [], []

for a in alphas:
    model = LogisticRegression(penalty='l1', solver='liblinear', C=a).fit(X, y)
    models.append(model)
    preds.append(model.coef_)
    
plt.figure(figsize=(12, 12))
for i in np.vstack(preds).T:
    plt.plot(range(len(preds)), i)

plt.ylim(-0.1, 0.15) 
plt.legend(columns, loc=5)      
plt.grid()
```

WOE_action_type опять впереди. Порядок остальных переменных немного отличается, но в целом остается в той же логике, что и при использовании Lasso: временные переменные выходят на лидирующие позиции, несмотря на низкие корреляции с таргетом.

```python
zeros_count = []

for arr in np.vstack(preds).T:
    zeros_count.append(list(arr).count(0))
    
columns_value['logit'] = zeros_count

columns_value.sort_values(by='logit')
```

```python
# PCA
pca = PCA(n_components=37).fit(X)
columns_value['PCA'] = pca.explained_variance_ratio_
columns_value.sort_values(by='PCA', ascending=False)
```

PCA пользуется совершенно другими принципами. WOE_action_type ушел куда-то в середину рейтинга. Самой значимой переменной оказалась action_type_Pullup Jump shot, второй по важности - month (то есть время года важнее, чем расстояние до корзины?)

```python
#Обучение моделей
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
logistic = LogisticRegression(penalty='l1', solver='liblinear').fit(X_train, y_train)
logistic.coef_[0]
```

```python
logistic_predicted = logistic.predict(X_train)
roc_auc_score(y_train, logistic_predicted)
```

```python
logistic_predicted = logistic.predict(X_test)
roc_auc_score(y_test, logistic_predicted)
```

```python
ridge = RidgeClassifier().fit(X_train, y_train)
ridge.coef_
```

```python
ridge_predicted = ridge.predict(X_train)
roc_auc_score(y_train, ridge_predicted)
```

```python
ridge_predicted = ridge.predict(X_test)
roc_auc_score(y_test, ridge_predicted)
```

```python
lasso = Lasso(alpha=0.0001).fit(X, y)
lasso.coef_
```

```python
lasso_predicted = lasso.predict(X_train)
roc_auc_score(y_train, lasso_predicted)
```

```python
lasso_predicted = lasso.predict(X_test)
roc_auc_score(y_test, lasso_predicted)
```

```python
#Интересно, что когда зануляются все коэффициенты, кроме WOE_action_type, lasso не намного хуже отрабатывает.
lasso = Lasso(alpha=0.1).fit(X, y)
list(zip(columns, lasso.coef_))
```

```python
lasso_predicted = lasso.predict(X_train)
roc_auc_score(y_train, lasso_predicted)
```

```python
lasso_predicted = lasso.predict(X_test)
roc_auc_score(y_test, lasso_predicted)
```

Все линейные модели отработали премерно одинаково. Лучшие результаты у Lasso, хотя эта модель и не предназначена для классификации. Настройки штрафа у ridge и logit очень мало влияли на значение ошибки, а в lasso уменьшение штрафа очень влияло на коэффициенты, и в итоге получилось подобрать alpha, которое дает максимальную площадь под кривой.
