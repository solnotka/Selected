Получаю переменные, связанные с количеством секунд в игре. Для этого у меня есть три столбца: period, minutes_remaining и seconds_remaining.
1. Значения в столбце seconds_remaining варьируются от 0 до 60. Значит, чтобы получить время до конца периода, нужно складывать minutes_remaining и seconds_remaining
2. В баскетбольном матче 4 периода по 12 минут. В датасете есть отметки о семи периодах. Предполагаю, что периоды 5-7 - это дополнительное время (по пять минут каждый).

```python
data[['period', 'minutes_remaining']].drop_duplicates()[data.period.isin([5, 6, 7])].describe()
```

```python
#количество секунд до конца периода
def time_remaining(row):
    return row['minutes_remaining'] * 60 + row['seconds_remaining']

#количество секунд, которое прошло с начала матча
def seconds_before(row):
    period = row['period']
    if period < 5:
        period_time = 12 * 60
        time_behind = period_time * (period - 1)
    else:
        period_time = 5 * 60
        time_behind = 12 * 60 * 4
        for i in range(5, period):
            time_behind += 300
    
    remaining = time_remaining(row)
    time_behind += (period_time - remaining)
    return time_behind

#количество секунд, которое прошло с начала периода
def period_seconds_before(row):
    period = row['period']
    if period < 5:
        period_time = 12 * 60
    else:
        period_time = 5 * 60
    
    remaining = time_remaining(row)
    time_behind = period_time - remaining
    return time_behind    
```

```python
data['all_seconds_remaining'] = data.apply(time_remaining, axis=1)
data['time_behind'] = data.apply(seconds_before, axis=1)
data['period_time_behind'] = data.apply(period_seconds_before, axis=1)
```

Сейчас у меня есть три категориальные переменные, потенциально очень значимые, которые надо подготовить для анализа: action_type, combined_shot_type и opponent. План работы с ними следующий:
1. Создаю столбцы с WOE и IV для этих признаков Буду считать эти показатели на data_train.
2. WOE оставлю как замену категориальных значений
3. IV использую для one hot encoding: буду создавать бинарные столбцы только для значений, у которых IV больше определенного порога
4. Устанавливаю минимальное значение IV в 0.01, чтобы точно не потерять важные признаки.

```python
def goods_bads(row, column):
    goods = len(data_train[data_train['shot_made_flag'] == 1])
    bads = len(data_train[data_train['shot_made_flag'] == 0])
    
    meaning_goods = len(data_train[(data_train['shot_made_flag'] == 1) & (data_train[column] == row[column])])
    meaning_bads = len(data_train[(data_train['shot_made_flag'] == 0) & (data_train[column] == row[column])])
    
    p_goods = meaning_goods / goods
    p_bads = meaning_bads / bads
    
    return p_goods, p_bads

def shot_WOE(row, column):
    p_goods, p_bads = goods_bads(row, column)  
    try:
        WOE =  np.log(p_goods / p_bads)
    except ZeroDivisionError:
        WOE = 1
    
    return WOE

def shot_IV(row, column):
    p_goods, p_bads = goods_bads(row, column)
    try:
        IV = (p_goods - p_bads) * np.log(p_goods / p_bads)
    except ZeroDivisionError:
        IV = p_goods - p_bads
    
    return IV
```

```python
data['WOE_combined_shot_type'] = data.apply(shot_WOE, axis = 1, column='combined_shot_type')
```

```python
data['IV_combined_shot_type'] = data.apply(shot_IV, axis = 1, column='combined_shot_type')
```

```python
#Генерация бинарных признаков
def pet_dummies(row, column, meaning):
    if row[column] == meaning:
        return 1
    else:
        return 0
```

```python
def generate_that(column, iv_column, name_part):
    for value in data[column].unique():
        IV = float(data.loc[:, iv_column][data[column] == value].drop_duplicates())
        if IV > 0.01:
            name = name_part + value
            data[name] = data.apply(pet_dummies, column=column, meaning=value, axis=1)
```
